## 0. Financial Data Acquisition: Global X Robotics & AI ETF (BOTZ)

To represent the market performance of the Artificial Intelligence sector, we utilize data from the **BOTZ ETF**, which tracks companies involved in robotics and AI. This step involves fetching high-fidelity historical data directly from the market.

### Technical Parameters:
* **Ticker:** BOTZ (Global X Robotics & Artificial Intelligence ETF).
* **Timeframe:** September 2016 to January 2024 (Aligning with the AI research data).
* **Metric:** Adjusted Daily Closing Prices.
* **Derived Variable:** Daily Returns (Percentage change) to measure market volatility and reaction speed.

This financial dataset will be used as the dependent variable in our correlation analysis to see how market movements synchronize with academic innovation.

```python
import yfinance as yf
import pandas as pd

# ETF ticker
ticker = "BOTZ"

# Fetching data from late 2016 to early 2024 as specified in the project proposal
botz = yf.Ticker(ticker)
data = botz.history(start="2016-09-01", end="2024-01-01")

# Extract only the closing price
df = data[["Close"]].copy()

# Rename the column for clarity
df.columns = ["price"]

# Calculate daily percentage returns
df["daily_return"] = df["price"].pct_change()

# Clean up index (yfinance history includes timezone info, we remove it for consistency)
df.index = pd.to_datetime(df.index).tz_localize(None)
df.index.name = "date"
df = df.reset_index()

# Select final columns and drop the first row (which will have a NaN daily return)
df = df[["date", "price", "daily_return"]]
df = df.dropna()

# Round values for clean data presentation
df = df.round({"price": 5, "daily_return": 5})

# Save to CSV for persistent storage
df.to_csv("botz_daily_data.csv", index=False)

print("Financial dataset successfully created and filtered until early 2024!")
print(df.tail()) # Verifying end dates with tail()
```
